```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Données
lacs = ['Lac Agnes', 'Lac Waterfowl', 'Lac Bow', 'Lac Peyto']
ph = [8.53, 8.34, 8.30, 8.29]
conductivite = [163.3, 182, 162.8, 151.4]
oxygene_dissous = [101.5, 103, 138.6, 106]
temperature = [11.4, 11.9, 9.9, 9.9]

# Style scientifique moderne
sns.set_theme(style="whitegrid")

# Création de graphiques séparés pour chaque paramètre
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# pH
axs[0, 0].bar(lacs, ph, width=0.4, color=sns.color_palette("deep")[0], edgecolor='black')
axs[0, 0].set_title('pH des lacs', fontsize=24, fontweight='bold')
axs[0, 0].set_ylabel('pH', fontsize=22)
axs[0, 0].set_xticklabels(lacs, fontsize=18)
axs[0, 0].grid(visible=True, linestyle='--', linewidth=0.7, alpha=0.7)
axs[0, 0].spines['top'].set_visible(False)
axs[0, 0].spines['right'].set_visible(False)
axs[0, 0].tick_params(axis='y', labelsize=18)  # Increase y-tick labels size

# Conductivité
axs[0, 1].bar(lacs, conductivite, width=0.4, color=sns.color_palette("deep")[1], edgecolor='black')
axs[0, 1].set_title('Conductivité des lacs', fontsize=24, fontweight='bold')
axs[0, 1].set_ylabel('Conductivité (µS/cm)', fontsize=22)
axs[0, 1].set_xticklabels(lacs, fontsize=18)
axs[0, 1].grid(visible=True, linestyle='--', linewidth=0.7, alpha=0.7)
axs[0, 1].spines['top'].set_visible(False)
axs[0, 1].spines['right'].set_visible(False)
axs[0, 1].tick_params(axis='y', labelsize=18)  # Increase y-tick labels size

# Oxygène dissous
axs[1, 0].bar(lacs, oxygene_dissous, width=0.4, color=sns.color_palette("deep")[2], edgecolor='black')
axs[1, 0].set_title('Oxygène dissous des lacs', fontsize=24, fontweight='bold')
axs[1, 0].set_ylabel('Oxygène dissous (mg/L)', fontsize=22)
axs[1, 0].set_xticklabels(lacs, fontsize=18)
axs[1, 0].grid(visible=True, linestyle='--', linewidth=0.7, alpha=0.7)
axs[1, 0].spines['top'].set_visible(False)
axs[1, 0].spines['right'].set_visible(False)
axs[1, 0].tick_params(axis='y', labelsize=18)  # Increase y-tick labels size

# Température
axs[1, 1].bar(lacs, temperature, width=0.4, color=sns.color_palette("deep")[3], edgecolor='black')
axs[1, 1].set_title('Température des lacs', fontsize=24, fontweight='bold')
axs[1, 1].set_ylabel('Température (°C)', fontsize=22)
axs[1, 1].set_xticklabels(lacs, fontsize=18)
axs[1, 1].grid(visible=True, linestyle='--', linewidth=0.7, alpha=0.7)
axs[1, 1].spines['top'].set_visible(False)
axs[1, 1].spines['right'].set_visible(False)
axs[1, 1].tick_params(axis='y', labelsize=18)  # Increase y-tick labels size

# Ajustement de la mise en page
plt.tight_layout()

# Affichage des graphiques
plt.show()

# Sauvegarde des graphiques avec une haute résolution
fig.savefig('lacs_parameters.png', dpi=300)
```
