```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

lacs = ['Lac Agnes', 'Lac Waterfowl', 'Lac Bow', 'Lac Peyto']
ph = [8.53, 8.34, 8.30, 8.29]
conductivite = [163.3, 182, 162.8, 151.4]
oxygene_dissous = [101.5, 103, 138.6, 106]
temperature = [11.4, 11.9, 9.9, 9.9]
turbidite = [np.nan, 4.34, 1.72, 5.10]

sns.set_theme(style="whitegrid")

if np.all(np.isnan(turbidite)):
    fig, axs = plt.subplots(2, 2, figsize=(16, 12))
else:
    fig, axs = plt.subplots(3, 2, figsize=(16, 18))

# pH
axs[0, 0].bar(lacs, ph, width=0.4, color=sns.color_palette("deep")[0], edgecolor='black')
axs[0, 0].set_title('pH des lacs', fontsize=28, fontweight='bold')
axs[0, 0].set_ylabel('pH', fontsize=26)
axs[0, 0].set_xticklabels(lacs, fontsize=18)
axs[0, 0].grid(visible=True, linestyle='--', linewidth=0.7, alpha=0.7)
axs[0, 0].spines['top'].set_visible(False)
axs[0, 0].spines['right'].set_visible(False)
axs[0, 0].tick_params(axis='y', labelsize=22)  # Increase y-tick labels size

# Conductivité
axs[0, 1].bar(lacs, conductivite, width=0.4, color=sns.color_palette("deep")[1], edgecolor='black')
axs[0, 1].set_title('Conductivité des lacs', fontsize=28, fontweight='bold')
axs[0, 1].set_ylabel('Conductivité (µS/cm)', fontsize=26)
axs[0, 1].set_xticklabels(lacs, fontsize=18)
axs[0, 1].grid(visible=True, linestyle='--', linewidth=0.7, alpha=0.7)
axs[0, 1].spines['top'].set_visible(False)
axs[0, 1].spines['right'].set_visible(False)
axs[0, 1].tick_params(axis='y', labelsize=22)  # Increase y-tick labels size

# Oxygène dissous
axs[1, 0].bar(lacs, oxygene_dissous, width=0.4, color=sns.color_palette("deep")[2], edgecolor='black')
axs[1, 0].set_title('Oxygène dissous des lacs', fontsize=28, fontweight='bold')
axs[1, 0].set_ylabel('Oxygène dissous (mg/L)', fontsize=26)
axs[1, 0].set_xticklabels(lacs, fontsize=18)
axs[1, 0].grid(visible=True, linestyle='--', linewidth=0.7, alpha=0.7)
axs[1, 0].spines['top'].set_visible(False)
axs[1, 0].spines['right'].set_visible(False)
axs[1, 0].tick_params(axis='y', labelsize=22)  # Increase y-tick labels size

# Température
axs[1, 1].bar(lacs, temperature, width=0.4, color=sns.color_palette("deep")[3], edgecolor='black')
axs[1, 1].set_title('Température des lacs', fontsize=28, fontweight='bold')
axs[1, 1].set_ylabel('Température (°C)', fontsize=26)
axs[1, 1].set_xticklabels(lacs, fontsize=18)
axs[1, 1].grid(visible=True, linestyle='--', linewidth=0.7, alpha=0.7)
axs[1, 1].spines['top'].set_visible(False)
axs[1, 1].spines['right'].set_visible(False)
axs[1, 1].tick_params(axis='y', labelsize=22)  # Increase y-tick labels size

# Turbidité
if not np.all(np.isnan(turbidite)):
    axs[2, 0].bar(lacs, turbidite, width=0.4, color=sns.color_palette("deep")[4], edgecolor='black')
    axs[2, 0].set_title('Turbidité des lacs', fontsize=28, fontweight='bold')
    axs[2, 0].set_ylabel('Turbidité (NTU)', fontsize=26)
    axs[2, 0].set_xticklabels(lacs, fontsize=18)
    axs[2, 0].grid(visible=True, linestyle='--', linewidth=0.7, alpha=0.7)
    axs[2, 0].spines['top'].set_visible(False)
    axs[2, 0].spines['right'].set_visible(False)
    axs[2, 0].tick_params(axis='y', labelsize=22)  # Increase y-tick labels size


    fig.delaxes(axs[2, 1])
    
    plt.subplots_adjust(hspace=0.3, wspace=0.3)
    
 
    pos = axs[2, 0].get_position()
    axs[2, 0].set_position([pos.x0 - 0.1, pos.y0, pos.width, pos.height])
else:
    fig.delaxes(axs[2, 0])
    fig.delaxes(axs[2, 1])


plt.tight_layout()


plt.show()


fig.savefig('lacs_parameters.png', dpi=300)
```

```python
import pandas as pd
from matplotlib.table import Table
import matplotlib.pyplot as plt

# Create the data directly
data = {
    'Lac': ['Lac Agnes', 'Lac Waterfowl', 'Lac Bow', 'Lac Peyto'],
    'pH': [8.53, 8.34, 8.30, 8.29],
    'Conductivité (µS/cm)': [163.3, 182, 162.8, 151.4],
    'Oxygène dissous (mg/L)': [101.5, 103, 138.6, 106],
    'Température (°C)': [11.4, 11.9, 9.9, 9.9],
    'Turbidité (NTU)': [np.nan, 4.34, 1.72, 5.10]
}
df = pd.DataFrame(data)

# Font parameters
font_size = 80  # Further increased font size
font_name = 'Arial'  # Optional: Specify a clear font

# Create a large figure to accommodate bigger text
fig, ax = plt.subplots(figsize=(20, 12), dpi=300)  # Increased figsize
ax.axis('off')

# Prepare table data with headers
table_data = [
    ['Lac', 'pH', 'Conductivité\n(µS/cm)', 'Oxygène\ndissous\n(mg/L)', 'Température\n(°C)', 'Turbidité\n(NTU)']
] + df.values.tolist()

# Create the table with more space
table = Table(ax, bbox=[0, 0, 1, 1])

# Define relative column widths
column_widths = [0.15, 0.1, 0.15, 0.15, 0.15, 0.15]

# Add cells to the table
for i, row in enumerate(table_data):
    for j, cell_value in enumerate(row):
        cell = table.add_cell(
            i, j, 
            width=column_widths[j],
            height=0.1,  # Adjust height as needed
            text=str(cell_value),
            loc='center',
            facecolor='white'
        )
        
        # Set font properties
        cell_text = cell.get_text()
        cell_text.set_fontsize(font_size)
        cell_text.set_fontname(font_name)
        
        # Styling for header row
        if i == 0:
            cell.set_facecolor('#cccccc')
            cell_text.set_weight('bold')
            cell_text.set_ha('center')
            cell_text.set_va('center')
            # Optional: Enable text wrapping if needed
            # cell_text.set_wrap(True)

# Add the table to the axes
ax.add_table(table)

# Adjust layout to fit the table
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05)

# Display the table
plt.show()
```
